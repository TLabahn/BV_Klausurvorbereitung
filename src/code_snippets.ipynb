# MuML/BV Code Schnipsel

```python
# Bibliotheken einbinden
import cv2
import sys
import matplotlib.pyplot as plt
```

## Mathematische Operationen / Elementare Operationen

```python
file = "../img/mandril_gray.png"
img = cv2.imread(file)

rows, columns, _ = img.shape
mean = 0.0

for row in range(rows):
    for column in range(columns):
        mean += img[row, column]
mean = mean / (rows * columns)

print(f"mean: {mean}")
```

## Bilder anzeigen

Mit OpenCV:

```python
file = "../img/mandril_gray.png"
img = cv2.imread(file)

if img is None:
    sys.exit("Could not read the image.")

# Anzeigen im Fenster
cv2.imshow('Monke', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Mit Matplotlib:

```python
file = "../img/mandril_gray.png"
img = cv2.imread(file)
# img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # OpenCV ldt BGR, Matplotlib erwartet RGB

plt.imshow(img)
plt.title('Monke')
plt.axis('off')
plt.show()
```

Mit Matplotlib und `plt.imread`:

```python
file = "../img/mandril_gray.png"
img = plt.imread(file)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # OpenCV ldt BGR, Matplotlib erwartet RGB

plt.imshow(img_rgb)
plt.title('Monke')
plt.axis('off')
plt.show()
```

## Farbbilder

```python
file = "../img/color_wheel.png"
img = cv2.imread(file)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # OpenCV ldt BGR, Matplotlib erwartet RGB

plt.imshow(img_rgb)
plt.title('Farbrad in Farbe')
plt.axis('off')
plt.show()
```

Explizit als Grauwertbild ffnen:

```python
# als Grauwertbild ffnen
img = cv2.imread(file, cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
plt.title('Farbrad als Grauwertbild')
plt.axis('off')
plt.show()
```
